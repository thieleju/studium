## Aufgabe 2: Plots mit Matplotlib erstellen

2.1 Importieren Sie Matplotlib.pyplot als plt

```python
import matplotlib.pyplot as plt
```

2.2 Plotten Sie die vorgegebenen Dictionaries in ein gemeinsames Liniendiagramm. Achten Sie auf eine sinnvolle Beschriftung der Achsen. Mehrfache Plots in einem Diagramm erhalten Sie durch mehrfache Anwendung der `plot()` Funktion.

Tip: 

- Die Werte für x als Liste erhalten Sie mit `list(my_dict.keys())`
- Die Werte für f(x) als Liste erhalten Sie mit `list(my_dict.values())`

```python
# Copied from solutions, because no dictionaries provided

# lambda for f(x)=2 * x^3 + x^2
# since range does not work with float, divide x value
# div value is 10 by default -> 30/10 = 3.0
f = lambda min,max,div=10 : {x/div: 2 * (x/div)**3 + (x/div)**2 for x in range(min, max)}
ff = lambda min,max,div=10 : {x/div: 2 * (x/div)*(3*(x/div)+1) for x in range(min, max)}

# generate f(x) and f'(x) values in range -6.0, 6.0
fy_val = f(-60, 60)
ffy_val = ff(-60, 60)
print(fy_val)
print(ffy_val)
```

```python
plt.plot(list(fy_val.keys()), fy_val.values())
plt.plot(list(ffy_val.keys()), ffy_val.values())
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Polynomfunktion")
plt.show()
```

2.3 Erzeugen Sie mit Numpy `randint` eine Liste aus Zufallszahlen und plotten diese als Graph

```python
np_list = np.random.randint(0,100,1000)

plt.plot(range(0,1000),np_list)
plt.show()
```

2.4 Sortieren Sie die Liste und plotten diese erneut als Graph

```python
x_list = np.sort(np_list)

plt.plot(range(0,1000),x_list)
plt.show()
```
