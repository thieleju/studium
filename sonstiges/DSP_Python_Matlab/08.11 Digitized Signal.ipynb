```python
"""
Digitized Signal
"""
import math
import numpy as np
import matplotlib.pyplot as plt

fs = 20
dt = 1 / fs
N = 64
```

```python
#%% Cell 1
k = np.arange(0,N)
t = k * dt
f_signal = 2
s = np.cos(2 * math.pi * f_signal * t)
```

```python
#%% Cell 2
plt.figure(1)
plt.subplot(211)
plt.plot(k,s,'k*-')
plt.grid()
plt.xlabel('Time index k')
plt.ylabel('s(t)')
plt.show()

plt.subplot(212)
plt.plot(t,s,'k*-')
plt.grid()
plt.xlabel('Time t in seconds')
plt.ylabel('s(t)')
plt.show()
```
