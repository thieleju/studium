```python
#%% Import libs
import math as ma
import numpy as np
import matplotlib.pyplot as plt

#%% Define variables
fs = 20
dt = 1/fs
N  = 64
k  = np.arange(0, N)
t  = k * dt
f_signal = 2

#%% Signal generation
s = np.cos(2 * ma.pi * f_signal * t)

#%% Plot Signal
plt.subplot(212)
plt.plot(k, s, "k*-")
plt.xlabel("Time index k")
plt.ylabel("s(k)")
plt.show()
```
