```matlab
%% First Program

%% Preparation
close all; clear all; clc; format compact;

%% Define values
dt =  0.02;
N = 256;
k = [0:1:N-1];
t = k * dt;
f = 1;

%% Calculate signal
s = sin(2 * pi * f * t);

%% plot 
figure;
plot(t,s,'*-');
xlabel("t in [s]")
ylabel("s(t)")
grid
```

```matlab
%% Createing a discrete signal in MATLAB

%% Define sampling frequency
fs=20;

%% Calculate time step
dt=1/fs;

%% Defined N and k and t
N=64;
k=0:1:N-1;
t=k*dt;

%% Defined frequency of signal
f_signal=2;

%% Calculate Signal and plot
s=cos(2*pi*f_signal*t);

subplot(2,1,1)
plot(k,s,'*-');

xlabel('Time index k')
ylabel('s(t)')

subplot(2,1,2)
plot(t,s,'*-');

xlabel('Time in seconds')
ylabel('s(t)')
```
