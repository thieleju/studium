### Operatoren und den resultierenden Datentypen
Wie schon in Übung 1 werden an dieser Stelle verschiedene Operationen durchgeführt und das Ergebnis mit entsprechndem Datentyp ausgegenen.

````python

def printWithType(a):
    print(f'{a:<35} | The result is of type {type(eval(a.rsplit("=", 1)[0])).__name__}')

# Hinweis auf Informationen aus Videomaterial:
# Nicht nur True und False können in Boolschen Operationen verwendet werden!
# Was ergibt in Python False:
#    False
#    None
#    Null Werte numerischer Typen: 0; 0.0; 0j
#    Empty Sequences, Tuples and Dicts: '', (), [], {}

# Boolsche Operatoren
# <, <=, ==, >=, >, !=, and, or, is
# Mehr zum 'is' Operator in playgroud_objects.py
printWithType(f'{False and True = }')
printWithType(f'{False or True = }')
printWithType(f'{0 < 1 = }')
printWithType(f'{0 >= 1 = }')
printWithType(f'{0 == 1 = }')
printWithType(f'{0 != 1 = }')

# In Operator auf Iterables
printWithType(f"{'Cat' in 'CatDog' = }")
printWithType(f"{'cat' in 'CatDog' = }")
printWithType(f"{'Cat' in ['Cat','Dog'] = }")
printWithType(f"{'cat' in ['Cat','Dog'] = }")
printWithType(f"{'Cat' in {'Cat':1,'Dog':4} = }")
printWithType(f"{'cat' in {'Cat':7,'Dog':5} = }")
printWithType(f"{None in ['Cat','Dog'] = }")
printWithType(f"{None in [None,'Dog'] = }")

# Bitwise Operatoren
printWithType(f'{False ^ True = }')
printWithType(f'{~False = }')
printWithType(f'{False >> 1 = }')
printWithType(f'{True << 1 = }')

# Bool-Werte (True, False) in numerischen Operationen -> Integer Werte (1 und 0)
printWithType(f'{1 * (1 == 10) = }')
printWithType(f'{4 / (1 == 1) = }')

# Numerische Typen können in Vergleichsoperationen verwendet werden, aber Achtung Genauigkeit
printWithType(f'{1 == 1.0000000000000001 = }')
printWithType(f'{1 == 1.000000000000001 = }')
printWithType(f'{1 == 1.0 == 1+0j = }')

# Fast alle Python Typen lassen sich in Boolschen Ausdrücken verwenden
printWithType(f'{1 and True = }')
printWithType(f'{1 and 1.9 = }')
printWithType(f'{True and 3j = }')
printWithType(f'{1 or True = }')
printWithType(f'{1.0 or True = }')
printWithType(f'{True or 1 = }')
printWithType(f'{None or None = }')
printWithType(f'{None or 1 = }')
printWithType(f'{0 and False = }')
printWithType(f'{False and 0 = }')
printWithType(f'{0 or False = }')
printWithType(f'{False or 0 = }')
printWithType(f'{[] or 0 and "" = }')
printWithType(f'{0 or [] and () = }')
printWithType(f'{[0] or 1 = }')
printWithType(f'{1 or [0] = }')

````

### Objekte aus der Speicherbrille
Wann werden Objekte mutiert und wann werden diese kopiert?
Für die Vergleichbarkeit sind jeweilige Speicheradressen farbig markiert. Sie können im Code schauen was gemacht wird und in der Ausgabe die Veränderung der Werte zusammen mit der Speicheradresse ansehen. Eigene Objekte können analog zur definition von Objekten in `objects_to_observe` angelegt und betrachtet werden.

````python
cCodes = \
    [
        '\033[31m',
        '\033[93m',
        '\033[33m',
        '\033[34m',
        '\033[35m',
        '\033[36m',
        '\033[94m',
        '\033[38m',
        '\033[92m',
        '\033[91m'
    ]


def color_int_string(some_string: str):
    return ''.join([cCodes[int(c)] + c for c in some_string]) + '\033[0m'


def idc(any_type: any):
    return color_int_string(str(id(any_type)))


def print_obj_information(obj_dict):
    for k, v in obj_dict.items():
        print(f'{k}@{idc(v)} {type(v)} is {v}')


list_of_ints = [1, 2, 3, 4]
objects_to_observe = \
    {
        'list_of_ints_1': list_of_ints,
        'list_of_ints_2': list_of_ints
    }
print('\nInitial data, two variables do contain the same data at the same memory address')
print_obj_information(objects_to_observe)

print('\nSort object _1 reverse → both content changes, address same')
objects_to_observe['list_of_ints_1'].sort(reverse=True)

print_obj_information(objects_to_observe)

print('\nAssign an equal list of numbers to _2 → address of _2 changes, content is still the same')
objects_to_observe['list_of_ints_2'] = [4, 3, 2, 1]

print_obj_information(objects_to_observe)

print('\nSort _1 again → only content _1 changes, ass only the contnet od address _1 is changed')
objects_to_observe['list_of_ints_1'].sort()

print_obj_information(objects_to_observe)

print('\nFun with slices')
objects_to_observe['slice_from_ints'] = objects_to_observe['list_of_ints_1'][1:3]
objects_to_observe['slice_from_ints_elem_1'] = objects_to_observe['slice_from_ints'][0]
objects_to_observe['list_of_ints_1_elem_1'] = objects_to_observe['list_of_ints_1'][1]

print_obj_information(objects_to_observe)

print('\nSlices are just lists with shallow copies')
objects_to_observe['list_of_ints_1'][1] = 10
objects_to_observe['list_of_ints_1'][1:3][1] = 20
objects_to_observe['slice_from_ints_elem_1_old'] = objects_to_observe['slice_from_ints'][0]

print_obj_information(objects_to_observe)


objects_to_observe.clear()
var_int = 4 if not objects_to_observe else 0
objects_to_observe['int_1'] = 4
objects_to_observe['int_2'] = objects_to_observe['int_1']
objects_to_observe['int_3'] = 4 + 1
objects_to_observe['int_4'] = 5
objects_to_observe['int_5'] = 1 + int(f'{var_int}')
objects_to_observe['int_6'] = 1 + int(f'{2+2}')
objects_to_observe['str_1'] = 'a'
objects_to_observe['str_2'] = 'b'
objects_to_observe['str_3'] = 'a'
objects_to_observe['str_4'] = f'{5+2}'
objects_to_observe['str_5'] = f'{5+2}'
objects_to_observe['str_6'] = f'7'
objects_to_observe['str_7'] = '7'

print(objects_to_observe['str_4'] == objects_to_observe['str_5']
      == objects_to_observe['str_6'] == objects_to_observe['str_7'])
print(objects_to_observe['str_4'] is objects_to_observe['str_5'])
print(objects_to_observe['str_5'] is objects_to_observe['str_6'])
print(objects_to_observe['str_6'] is objects_to_observe['str_7'])

print_obj_information(objects_to_observe)
````
