```

```
Überprüfen Sie das Ergebnis mit der Ausgabe des `**`-Operators

Aufgabe 2.4 (Bedingungen) | Bestimmen Sie ob ein angegebenes Jahr ein Schaltjahr ist.
```
Es gelten folgende Regeln:
  nicht durch 4 teilbar            → kein Schaltjahr
  durch 4 teilbar                  → Schaltjahr
  außer wenn durch 100 teilbar     → kein Schaltjahr
  außer wenn durch 400 teilbar     → Schaltjahr
```
```
Test:
  1900 → kein Schaltjahr
  2000 → Schaltjahr
  2004 → Schaltjahr
  2006 → kein Schaltjahr
```

````python
test_cases = { 1900: False, 2000: True, 2004: True, 2006: False, 2100: False, 2400: True}

# Function 1
def is_leap_year(year):
    if year % 4 == 0:
        if year % 100 == 0:
            if year % 400 == 0:
                return True
            return False
        return True
    return False
  
# Function 2
def is_leap_year_2(year):
    return year % 4 == 0 and (year % 100 != 0 or year % 400 == 0)

# Tests
def run_tests(test_cases, func):
    results = {"Success": 0, "Fail": 0}
    for year, expected in test_cases.items():
        if func(year) == expected:
            results["Success"] += 1
        else:
            results["Fail"] += 1
    print(f"Testing {func.__name__}: {'Success' if results['Fail'] == 0 else 'Fail'} ({results['Success']}/{len(test_cases)})")
    

run_tests(test_cases, is_leap_year)
run_tests(test_cases, is_leap_year_2)
````
