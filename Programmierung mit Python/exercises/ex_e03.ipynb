Aufgabe 1.1 (Imports, Funktionen) | Schreiben Sie eine Funktion `berechne_rueckgeld` die das Rückgeld an den Kunden im Supermarkt berechnet. Zur Einfachheit entstehen beim Bezahlen und beim Einkaufspreis keine Cent-Beträge. Ebenfalls soll immer nur Münzgeld zurückgegeben werden.
```
Input:
  preis(int):  der Preis der von Kunden zu zahlen ist
  gegeben(int): den gegebenen Betrag des Kunden

Beschreibung:
  Die Funktion soll die geringste mögliche Anzahl von Euromünzen zurückgeben, die das Rückgeld darstellen.

Output:
  dictionary: mit Anzahl der Münzen einer Wähung
```
```
Beispiel:
  preis:   13
  gegeben: 20
  return: {'2€': 3, '1€': 1}
```
Überlegen Sie sich was passieren soll, wenn 'preis' > 'gegeben' an die Funktion übergeben wird oder einer der Argumente negativ ist.

````python
def berechne_rueckgeld(preis: int, gegeben: int):
    if preis > gegeben:
        raise ValueError("Der gegebene Betrag ist nicht ausreichend, um den Preis zu decken.")
    if preis < 0 or gegeben < 0:
        raise ValueError("Der Preis und der gegebene Betrag dürfen nicht negativ sein.")

    rueckgeld = gegeben - preis
    rueckgabe = {'2€': 0, '1€': 0}

    if rueckgeld >= 2:
        rueckgabe['2€'] = rueckgeld // 2
        rueckgeld %= 2

    if rueckgeld >= 1:
        rueckgabe['1€'] = rueckgeld

    return rueckgabe

# Test der Funktion
print(berechne_rueckgeld(13, 20))
print(berechne_rueckgeld(14, 20))
print(berechne_rueckgeld(20, 20))
try:
    print(berechne_rueckgeld(20, -13))
except Exception as e:
    print(e)
try:
    print(berechne_rueckgeld(-20, -13))
except Exception as e:
    print(e)
try:
    print(berechne_rueckgeld(21, 20))
except Exception as e:
    print(e)
````

Aufgabe 1.2 (Imports, Funktionen) | Schreiben Sie eine Funktion `get_upper_rnd`.
Diese nimmt als Input einen positiven Integer `k` und gibt einen zufälligen Integer Wert `x` mit `k//2 <= x <= k` zurück.
Verwenden Sie das `random` Modul.

````python
import random

def get_upper_rnd(k):
    return random.randint(k // 2, k)
````

Aufgabe 1.3 (Imports, Funktionen) | Vervollständigen Sie die Funktion `time_it`.
Diese Funktion nimmt einen String `func_description` und eine parameterlose Funktion `func` als Input.

`time_it` soll die Funktion `func` ausführen, derern Laufzeit messen und das Ergebnis ausgeben.

Nutzen Sie das `time` Modul.
Welche Einheit hat die gemessene Zeit (`h,min,s,ms,µs,ns`)?

Hinweis: Für das Messen von Zeiten stehen eine Vielzahl an Möglichkeiten zur Verfügung
Lesen Sie auch über den Build-In `timeit`-Befehl

````python
import time

def time_it(f_description, func):
    # Start time measurement
    start_time = time.time() 
    func()
    # End time measurement
    end_time = time.time()
    elapsed_time = end_time - start_time
    # Print the result
    print(f"{f_description}: {elapsed_time} seconds.")
    
    
# test
def sleep_one_second():
  time.sleep(1)
time_it('Diese Funktion benötigt etwas eine Sekunde zum Ausführen.', sleep_one_second)
````
