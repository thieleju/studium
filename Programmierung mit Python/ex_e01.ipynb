## Playground
Hier gibt es nochmal weiterführende Details zu Operatoren und den resultierenden Datentypen.

````python
def printWithType(a):
    print(
        f'{a:<30} | The result is of type {type(eval(a.split(" = ")[0])).__name__}')

print('\nTest verschiedener Operatoren und den resultierenden Datentypen:')
printWithType(f'{123 + 321 = }')
printWithType(f'{1 - 0.0 = }')
printWithType(f'{-0 + 1 = }')
printWithType(f'{3 // 2 = }')
printWithType(f'{3.2 // 2 = }')
printWithType(f'{3 % 2 = }')
printWithType(f'{3 ** 2 = }')
printWithType(f'{4 ** 0.5 = }')
printWithType(f'{"1" + "2" = }')
printWithType(f'{2 / 1 = }')

print('\nWeitere Zahlenrepräsentation und Datenformate:')
printWithType(f'{.11 = }')
printWithType(f'{11. = }')
printWithType(f'{11e+3 = }')
printWithType(f'{-.3e-3 = }')
printWithType(f'{6j = }')
printWithType(f'{0xFF = }')
printWithType(f'{0o77 = }')
printWithType(f'{0b01010101 = }')
printWithType(f'{0xA0F + 0o37 - 0b1 - 2600 = }')

print('\nBoolsche Ausdrücke in Python:')
printWithType(f'{True = }')
printWithType(f'{False = }')
printWithType(f'{bool(1.0) = }')
printWithType(f'{bool(0.1) = }')
printWithType(f'{bool(0.0) = }')
printWithType(f'{bool(0) = }')
# mit None-Values, be careful with None
printWithType(f'{bool(None) = }')
printWithType(f'{None == False = }') # be careful with None
printWithType(f'{None == True = }')  # be careful with None
printWithType(f'{None == None = }')  # be careful with None
printWithType(f'{None != None = }')  # be careful with None
# mit Lists, Dicts und Sets
printWithType(f'{bool([]) = }') 
printWithType(f'{bool([False]) = }')
printWithType(f'{bool({}) = }')
printWithType(f'{bool({False}) = }') 
printWithType(f'{bool({False:False}) = }') 


print('\nSets und Dicts:')
ab = {'a', 'b'}
bc = {'b', 'c'}
printWithType(f'{ab - bc = }')
printWithType(f'{bc - ab = }')
printWithType(f'{ab | bc = }')
printWithType(f'{ab ^ bc = }')
printWithType(f'{ab & bc = }')

print('\nRepräsentation des Types Float als String:')
print(f'{1e-3}')
print(f'{1e-11}')
print(f'{1e-11:.3f}')
print(f'{1e-11:.11f}')
print(f'{0.12345678901234567890123456789}')
print(f'{0.12345678901234567890123456789:.64f}')
print(f'{1234567890.12345678901234567890123456789:.64f}')
````
