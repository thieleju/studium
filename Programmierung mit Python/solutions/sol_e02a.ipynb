## Aufgabe 1
Die erste Aufgabe beschäftigt sich mit Dictionaries und Loops.

Aufgabe 1.1 (Dicts, Loops) | Gegeben ist die variable `wiki_text` mit Wikipedia Text aus dem Einleitungsabschnitt 'Maschinelles Lernen'.

Befüllen Sie die Variable `char_frequency` vom Typ Dictionary.
In `char_frequency` sollen Buchstaben-Zahlen-Paare enthalten sein.
Die Buchstaben ergeben sich aus allen alphabetischen Buchstaben aus der Zeichenkette `wiki_text`.
Dabei soll nicht zwischen Groß- und Kleinschreibung unterschieden werden.
Die Zahl in dem Buchstaben-Zahlen-Paar gibt die Häufigkeit des Buchstabens in der Zeichenkette `wiki_text` an.
```
Beispiel: char_frequency = {'a': 12, 't': 6, 'c': 3} 
Dieses Dictonary gäbe an, dass der Buchstabe 'a' 12-mal, 
der Buchstabe 't' 6-mal und der Buchstabe 'c' 3-mal 
in einer Zeichenkette wiki_text enthalten ist.
```

````python
wiki_text = '''
Maschinelles Lernen ist ein Oberbegriff fuer die kuenstliche Generierung von Wissen aus Erfahrung: 
Ein kuenstliches System lernt aus Beispielen und kann diese nach Beendigung der Lernphase verallgemeinern. 
Dazu bauen Algorithmen beim maschinellen Lernen ein statistisches Modell auf, das auf Trainingsdaten beruht. 
Das heisst, es werden nicht einfach die Beispiele auswendig gelernt, sondern Muster und Gesetzmaessigkeiten 
in den Lerndaten erkannt. So kann das System auch unbekannte Daten beurteilen (Lerntransfer) oder aber am 
Lernen unbekannter Daten scheitern (Ueberanpassung).
'''

char_frequency = {}
for c in wiki_text.lower():
    if c.isalpha():
        char_frequency[c] = char_frequency.get(c, 0) + 1

# Dieser Ausdruck sortiert 'char_frequency' nach den am häufigsten vorkommenden Buchstaben
char_frequency = dict(sorted(char_frequency.items(), key=lambda el: -el[1]))
print(char_frequency)
````

Es geht darum mit Hilfe der Nennung einzelner Buchstaben ein Wort zu erraten.

```
Regeln:

  - der Buchstabe schon mal geraten wurde
  - der Buchstabe nicht im Wort enthalten ist
  - das Wort nicht das Richtige ist
  Das Spiel endet, wenn:
  - der Spieler seine Versuche aufgebraucht hat → verloren
  - der Spieler das Wort vollständig erraten hat → gewonnen
```

````python
##### User settings #####
allowed_fails = 7
guess_word_user = 'Schmetterling'
##### User settings #####

# Lösung 2.3: -> 4 fails (hier Lösung als Code)
#char_index = {k: i for i, (k, _) in enumerate(char_frequency.items())}
char_index = {}
i = 0
for k, v in char_frequency.items():
    char_index[k] = i
    i += 1
# unique_chars = {c for c in guess_word_user.lower()}
unique_chars = set()
for c in guess_word_user.lower():
    unique_chars.add(c)
# last_index = max([char_index[c] for c in unique_chars])
last_index = 0
for c in unique_chars:
    if char_index[c] > last_index:
        last_index = char_index[c]
stupid_computer_fails = last_index - len(unique_chars) + 1

##### Game Variables #####
guess_word = guess_word_user.lower()       # lowercase word to play with
ties_left = allowed_fails                  # guesses left before lose
# Initialisieren Sie die Liste guessed_state mit lauter '_'-Buchstaben.
# Die Liste guessed_state enthält so viel Elemente wie guess_word
guessed_state = []                         # current guess state, e.g. '_ c h _ e t t e _ _ _ _ _'
for a in guess_word:                       # if 'c', 'h', 'e', 't' have been chosen
    guessed_state.append('_')
# Alterntive: guessed_state = ['_' for _ in guess_word]
                                           # if 'c', 'h', 'e', 't' have been chosen
guessed_chars = []                         # chars that have already been guessed
game_round = 0                             # rounds played
##### Game Variables #####

print('''
Welcome to Hangman in Python

''')
print(f'The stupid computer would have only made {stupid_computer_fails} false guesses to find the secret word.')
print('Can you beat him?\n')

while True:
    game_round += 1
    print(f'Round {game_round: 3d} | Current state {" ".join(guessed_state)} | {ties_left} tries left!')
    tip = input('\nChoose a letter or solve the word')
    print(f'\nYour tip is {tip}')

    if tip:
        l_tip = tip.lower()
        # Fall 1: User gibt ein Wort an und versucht zu lösen
        if len(l_tip) > 1:
            if l_tip == guess_word:
                guessed_state = []
                for a in guess_word:
                    guessed_state.append(a)
                # Alternative: guessed_state = [a for a in guess_word]
            else:
                ties_left -= 1
        # Fall 2: User gibt einen Buchstaben an
        else:
            if l_tip in guessed_chars:
                print(f'You already picked "{l_tip}" that one before loser!')
                ties_left -= 1
            else:
                if l_tip in guess_word:
                    for index, letter in enumerate(guess_word):
                        if letter == l_tip:
                            guessed_state[index] = l_tip
                    guessed_chars.append(l_tip)
                else:
                    ties_left -= 1

                # Alternative:
                # correct_indices = [x for x, letter in enumerate(guess_word) if letter == l_tip]
                # for index in correct_indices:
                #     guessed_state[index] = l_tip
                #
                # if correct_indices:
                #     guessed_chars.append(l_tip)
                # else:
                #     ties_left -= 1

        if '_' not in guessed_state:
            print(f'You Won! Word is "{guess_word_user}". | Tries left: {ties_left}')
            break
        elif not ties_left:
            print(f'GAME OVER! The word was "{guess_word_user}".')
            break
````

### Operatoren und den resultierenden Datentypen
Wie schon in Übung 1 werden an dieser Stelle verschiedene Operationen durchgeführt und das Ergebnis mit entsprechndem Datentyp ausgegenen.

````python

def printWithType(a):
    print(f'{a:<35} | The result is of type {type(eval(a.rsplit("=", 1)[0])).__name__}')

# Hinweis auf Informationen aus Videomaterial:
# Nicht nur True und False können in Boolschen Operationen verwendet werden!
# Was ergibt in Python False:
#    False
#    None
#    Null Werte numerischer Typen: 0; 0.0; 0j
#    Empty Sequences, Tuples and Dicts: '', (), [], {}

# Boolsche Operatoren
# <, <=, ==, >=, >, !=, and, or, is
# Mehr zum 'is' Operator in playgroud_objects.py
printWithType(f'{False and True = }')
printWithType(f'{False or True = }')
printWithType(f'{0 < 1 = }')
printWithType(f'{0 >= 1 = }')
printWithType(f'{0 == 1 = }')
printWithType(f'{0 != 1 = }')

# In Operator auf Iterables
printWithType(f"{'Cat' in 'CatDog' = }")
printWithType(f"{'cat' in 'CatDog' = }")
printWithType(f"{'Cat' in ['Cat','Dog'] = }")
printWithType(f"{'cat' in ['Cat','Dog'] = }")
printWithType(f"{'Cat' in {'Cat':1,'Dog':4} = }")
printWithType(f"{'cat' in {'Cat':7,'Dog':5} = }")
printWithType(f"{None in ['Cat','Dog'] = }")
printWithType(f"{None in [None,'Dog'] = }")

# Bitwise Operatoren
printWithType(f'{False ^ True = }')
printWithType(f'{~False = }')
printWithType(f'{False >> 1 = }')
printWithType(f'{True << 1 = }')

# Bool-Werte (True, False) in numerischen Operationen -> Integer Werte (1 und 0)
printWithType(f'{1 * (1 == 10) = }')
printWithType(f'{4 / (1 == 1) = }')

# Numerische Typen können in Vergleichsoperationen verwendet werden, aber Achtung Genauigkeit
printWithType(f'{1 == 1.0000000000000001 = }')
printWithType(f'{1 == 1.000000000000001 = }')
printWithType(f'{1 == 1.0 == 1+0j = }')

# Fast alle Python Typen lassen sich in Boolschen Ausdrücken verwenden
printWithType(f'{1 and True = }')
printWithType(f'{1 and 1.9 = }')
printWithType(f'{True and 3j = }')
printWithType(f'{1 or True = }')
printWithType(f'{1.0 or True = }')
printWithType(f'{True or 1 = }')
printWithType(f'{None or None = }')
printWithType(f'{None or 1 = }')
printWithType(f'{0 and False = }')
printWithType(f'{False and 0 = }')
printWithType(f'{0 or False = }')
printWithType(f'{False or 0 = }')
printWithType(f'{[] or 0 and "" = }')
printWithType(f'{0 or [] and () = }')
printWithType(f'{[0] or 1 = }')
printWithType(f'{1 or [0] = }')

````

### Objekte aus der Speicherbrille
Wann werden Objekte mutiert und wann werden diese kopiert?
Für die Vergleichbarkeit sind jeweilige Speicheradressen farbig markiert. Sie können im Code schauen was gemacht wird und in der Ausgabe die Veränderung der Werte zusammen mit der Speicheradresse ansehen. Eigene Objekte können analog zur definition von Objekten in `objects_to_observe` angelegt und betrachtet werden.

````python
cCodes = \
    [
        '\033[31m',
        '\033[93m',
        '\033[33m',
        '\033[34m',
        '\033[35m',
        '\033[36m',
        '\033[94m',
        '\033[38m',
        '\033[92m',
        '\033[91m'
    ]


def color_int_string(some_string: str):
    return ''.join([cCodes[int(c)] + c for c in some_string]) + '\033[0m'


def idc(any_type: any):
    return color_int_string(str(id(any_type)))


def print_obj_information(obj_dict):
    for k, v in obj_dict.items():
        print(f'{k}@{idc(v)} {type(v)} is {v}')


list_of_ints = [1, 2, 3, 4]
objects_to_observe = \
    {
        'list_of_ints_1': list_of_ints,
        'list_of_ints_2': list_of_ints
    }
print('\nInitial data, two variables do contain the same data at the same memory address')
print_obj_information(objects_to_observe)

print('\nSort object _1 reverse → both content changes, address same')
objects_to_observe['list_of_ints_1'].sort(reverse=True)

print_obj_information(objects_to_observe)

print('\nAssign an equal list of numbers to _2 → address of _2 changes, content is still the same')
objects_to_observe['list_of_ints_2'] = [4, 3, 2, 1]

print_obj_information(objects_to_observe)

print('\nSort _1 again → only content _1 changes, ass only the contnet od address _1 is changed')
objects_to_observe['list_of_ints_1'].sort()

print_obj_information(objects_to_observe)

print('\nFun with slices')
objects_to_observe['slice_from_ints'] = objects_to_observe['list_of_ints_1'][1:3]
objects_to_observe['slice_from_ints_elem_1'] = objects_to_observe['slice_from_ints'][0]
objects_to_observe['list_of_ints_1_elem_1'] = objects_to_observe['list_of_ints_1'][1]

print_obj_information(objects_to_observe)

print('\nSlices are just lists with shallow copies')
objects_to_observe['list_of_ints_1'][1] = 10
objects_to_observe['list_of_ints_1'][1:3][1] = 20
objects_to_observe['slice_from_ints_elem_1_old'] = objects_to_observe['slice_from_ints'][0]

print_obj_information(objects_to_observe)


objects_to_observe.clear()
var_int = 4 if not objects_to_observe else 0
objects_to_observe['int_1'] = 4
objects_to_observe['int_2'] = objects_to_observe['int_1']
objects_to_observe['int_3'] = 4 + 1
objects_to_observe['int_4'] = 5
objects_to_observe['int_5'] = 1 + int(f'{var_int}')
objects_to_observe['int_6'] = 1 + int(f'{2+2}')
objects_to_observe['str_1'] = 'a'
objects_to_observe['str_2'] = 'b'
objects_to_observe['str_3'] = 'a'
objects_to_observe['str_4'] = f'{5+2}'
objects_to_observe['str_5'] = f'{5+2}'
objects_to_observe['str_6'] = f'7'
objects_to_observe['str_7'] = '7'

print(objects_to_observe['str_4'] == objects_to_observe['str_5']
      == objects_to_observe['str_6'] == objects_to_observe['str_7'])
print(objects_to_observe['str_4'] is objects_to_observe['str_5'])
print(objects_to_observe['str_5'] is objects_to_observe['str_6'])
print(objects_to_observe['str_6'] is objects_to_observe['str_7'])

print_obj_information(objects_to_observe)
````
